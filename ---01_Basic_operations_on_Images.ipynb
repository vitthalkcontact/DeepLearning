```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## Basic operations on Images

####    Read & Writing an Image

```python
img = cv.imread("pylogo.png")
# img = img[::-1,:,::-1] # TO reverse color channels that will be plotted properly using matplot imshow()
cv.imshow("Image",img)

# Using cv API
cv.waitKey(0) # To get the key event
cv.destroyAllWindows() # To destory all the image GUI window

# Writing an image
cv.imwrite("pylog_write.png",img)
# Using matlab API
# plt.imshow(img)
# plt.show()
```

####    Access pixel properties, values & modifying

```python
img = cv.imread("pylogo.png")
# print(img)
print(img.shape)
print(img.dtype)
```

####    Splitting & Merging of image channels

```python
# Splitting an image
img = cv.imread("pylogo.png")
R,G,B = cv.split(img)
plt.subplot(131),plt.imshow(R)
plt.subplot(132),plt.imshow(G)
plt.subplot(133),plt.imshow(B)
plt.show()

# Merging an image
tmp = cv.merge([B,G,R])
plt.imshow(tmp)
plt.show()
```

####    Arithmetic Operation

```python
img1 = cv.imread("pylogo.png")
img2 = cv.imread("lena.png")
# tmp = img1
# tmp = img1+80
# tmp = img1+img2[:85,:85] # Adding two images, size must be same ; This is numpy addition with modulo
tmp = cv.add(img1,img2[:85,:85]) # Adding using cv.add(); This is saturated 
plt.imshow(tmp)
plt.show()
```

####    Bitwise Operation

```python
ones = np.ones([5,5])
zeros = np.zeros([5,5])
bit_and = cv.bitwise_and(ones,zeros) # 1*0 = 0, 1*1 = 1
bit_or = cv.bitwise_or(ones,zeros) # 1or0 = 1, 1or1 = 1
bit_xor = cv.bitwise_xor(ones,zeros) # 1xor0 = 1, 1xor1 = 0
ones_bit_not = cv.bitwise_not(ones) # not-1 = 0, not-0= 1
zeros_bit_not = cv.bitwise_not(zeros) # It makes zeros as nan which in turns treated as zero hence image is black
cv.imshow("Ones",ones)
cv.imshow("Zeros",zeros)
cv.imshow("bit_and",bit_and)
cv.imshow("bit_or",bit_or)
cv.imshow("ones_bit_not",ones_bit_not)
cv.imshow("zeros_bit_not",zeros_bit_not)
cv.waitKey(0) # To get the key event
cv.destroyAllWindows() # To destory all the image GUI window
```
