### GUI Features    

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

#### Display images in window

```python
img = cv.imread("pylogo.png")
# img = img[::-1,:,::-1] # TO reverse color channels that will be plotted properly using matplot imshow()
cv.imshow("Image",img)

# Using cv API
cv.waitKey(0) # To get the key event
cv.destroyAllWindows() # To destory all the image GUI window
```

#### Getting started with video capturing

```python
cap = cv.VideoCapture(0) # Generate an instance of VideoCapture(). 0 = webcam of PC/laptop. 
                         # Change the numnber based on number of camera installed in PC
while True:
    ret,frame = cap.read() # Will keep reading frames and Rettun true if frames available
    gray = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    cv.imshow("fame",frame)
    cv.imshow("fame_gray",gray)
    if cv.waitKey(1) & 0xFF == ord("q"):
        break
cap.release()
cv.destroyAllWindows()
```

#### Drawing Functions like circle, line, rectangle, polyline

```python
img = cv.imread("lena.png")

# Line
cv.line(img,(0,len(img)),(len(img),0),(255,0,0),2) # args are: image,point1(x,y) and point2(x,y) in the form of tuple
# and then RGB color combination then thickness of line

# Rectangle
cv.rectangle(img,(40,90),(90,40),(0,233,233),5) # args are: image,point1(x,y) and point2(x,y) in the form of tuple
# and then RGB color combination then thickness of line (if -1, then it will be filled else no fill)

# Filled Rectangle
cv.rectangle(img,(60,30),(30,50),(240,233,0),-1)

# Circle
cv.circle(img,(120,120),50,(0,255,0),10) # args are: image,point1(x,y) and point2(x,y) in the form of tuple
# and then RGB color combination then thickness of line (if -1, then it will be filled else no fill)

# Polygon
points = np.array([[0,40],[75,90],[90,60],[100,120],[120,200]])
cv.polylines(img,[points],True,(200,50,130),5) # args are: image,points in the form of array
# and then True (to close the loop) then BGR color combination then thickness of line (if -1, then it will be filled else no fill)

# Ellipse
cv.ellipse(img,(130,120),(55,20),0,0,360,(20,55,190),3) # args are: Plz chk the docstring

plt.figure(figsize=(6,6))
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.show()
```
